```python
#import the numpy module
import numpy as np

#declare an array with 50 random values...
a_giselle= np.random.randn(50)

#print the array...
print(a_giselle)

#find the mean of the array and print it...
mean_a = np.mean(a_giselle)
print("Mean of the array is = {:.3f}".format(mean_a))

#find the median of the array and print it...
median_a = np.median(a_giselle)
print("Median of the array is = {:.3f}".format(median_a))

#find the standard deviation of the array and print it...
std_a = np.std(a_giselle)
print("Standard deviation of the array is = {:.3f}".format(std_a))

#reshape the array
b_giselle = np.reshape(a_giselle, (5,10))
print("reshaped array b_giselle = \n\n",b_giselle)

c_giselle = np.reshape(a_giselle, (10,5))
print("reshaped array c_giselle = \n\n",c_giselle)

#print the element of b_giselle whose position is 3*7...
print("The element of array b_giselle at (3,7) is = {}".format(b_giselle[3][7]))

#print the sum of 8th row of the c_giselle array by using np.sum()...
print("The sum of the row 8 from array c_giselle = {}".format(np.sum(c_giselle[8])))
```

```python
import skimage
import matplotlib.pyplot as plt
from skimage import io, filters

# Load the data for coffee
coffee_giselle_fe = io.imread('https://github.com/scikit-image/scikit-image/blob/main/skimage/data/coffee.png?raw=true')

# Print the above data in textual format
print(coffee_giselle_fe)

# Print the data type and the data shape for the above data.
print("Data type:", type(coffee_giselle_fe))
print("Data shape:", coffee_giselle_fe.shape)

# Display the above data as an image using matplotlib; save the image as Figure1_first_last
plt.imshow(coffee_giselle_fe)
plt.savefig('Figure1_first_last.png')

# Import filters and use it to filter the above image using the Gaussian filter with std 5 (to blur it); save the image as filtered_Figure_first_last
coffee_filtered = filters.gaussian(coffee_giselle_fe, sigma=5)
plt.imshow(coffee_filtered)
plt.savefig('filtered_Figure_first_last.png')
```
